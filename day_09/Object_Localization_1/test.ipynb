```python
import csv

def find_row_with_value(csv_file_path, value):
    with open(csv_file_path, 'r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0] == value:
                return row
    return None  # If value is not found

# Find the row with the first value "image_0005"
value_to_find = img_path.split('/')[-1].replace('.jpg', '')
folder = img_path.split('/')[-2]

row = find_row_with_value('dataset/annotations/{}.csv'.format(folder), value_to_find)
```

```python
x1, y1, x2, y2 = row[1:-1]
```

```python
img = cv2.rectangle(img, (int(x1), int(y1)),
                    (int(x2), int(y2)), (255, 0, 0), 2)
plt.imshow(img[..., ::-1])
```
